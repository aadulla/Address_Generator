```python
import numpy as np
import copy
import math
import random

from Tester import Tester
```

```python
tester = Tester()
```

```python
loop_tiling_lst =   [ 
                        [{"weight": 2}, {"filter": 2}, {"channel": 2}, {"output": 2}],
                        [{"filter": 2}, {"output": 2}, {"channel": 2}, {"weight": 2}]
                    ]

# which data should write backs be enabled for
write_backs = {"input", "weight", "output"}
costs = [10, 5]
expansion_factor = 0

tester.setup(loop_tiling_lst, costs, expansion_factor, write_backs, debug=False)
tester.test(full=True)
```

```python
loop_tiling_lst =   [ 
                        [{"filter": 2}, {"channel": 2}, {"weight": 2}, {"output": 3}],
                        [{"filter": 2}, {"channel": 2}, {"weight": 2}, {"output": 2}]
                    ]

# which data should write backs be enabled for
write_backs = {"input", "output"}
costs = [10, 5]
expansion_factor = 0.5

tester.setup(loop_tiling_lst, costs, expansion_factor, write_backs, debug=False)
tester.test(full=True)
```
